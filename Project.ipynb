```python
""" Using sklearn make classification model """
# 1. Libraries import
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression

# 2. Data Prepare
from sklearn.datasets import load_digits
from sklearn.datasets import load_wine
from sklearn.datasets import load_breast_cancer
from Preprocess.DataPreprocess import Preprocess

# 3. Metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
```

# 손글씨 분류 모델

```python
X_train, X_test, Y_train, Y_test = Preprocess(load_digits(), split=True, test_size=0.2, random_state=16)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
# Model training and prediction

rs = 32

def classifier(DT=False, RF=False, SVM=False, SGD=False, LOG=False):
    if DT==True:
        decision_tree = DecisionTreeClassifier(random_state=rs)
        decision_tree.fit(X_train, Y_train)
        prediction1 = decision_tree.predict(X_test)
        accuracy1 = accuracy_score(Y_test, prediction1)
        cm = confusion_matrix(Y_test, prediction1)
        sns.heatmap(cm, annot=True, cmap="Reds")
        plt.xlabel('Label')
        plt.ylabel('Prediction')
        plt.show()
        return print("Accuracy:", np.round(accuracy1, 2), '\n' ,classification_report(Y_test, prediction1))
    
    else:
        pass

    if RF==True:
        random_forest = RandomForestClassifier(random_state=rs)
        random_forest.fit(X_train, Y_train)
        prediction2 = random_forest.predict(X_test)
        accuracy2 = accuracy_score(Y_test, prediction2)
        cm2 = confusion_matrix(Y_test, prediction2)
        sns.heatmap(cm2, annot=True, cmap="crest")
        plt.xlabel('Label')
        plt.ylabel('Prediction')
        plt.show()
        return print("Accuracy:", np.round(accuracy2, 2), '\n' ,classification_report(Y_test, prediction2))
    
    else:
        pass

    if SVM==True:
        svm_model = svm.SVC(random_state=rs)
        svm_model.fit(X_train, Y_train)
        prediction3 = svm_model.predict(X_test)
        accuracy3 = accuracy_score(Y_test, prediction3)
        cm3 = confusion_matrix(Y_test, prediction3)
        sns.heatmap(cm3, annot=True, cmap='Blues')
        plt.xlabel('Label')
        plt.ylabel('Prediction')
        plt.show()
        return print("Accuracy:", np.round(accuracy3, 2), '\n' ,classification_report(Y_test, prediction3))
    
    else:
        pass

    if SGD==True:
        sgd_model = SGDClassifier(random_state=rs)
        sgd_model.fit(X_train, Y_train)
        prediction4 = sgd_model.predict(X_test)
        accuracy4 = accuracy_score(Y_test, prediction4)
        cm4 = confusion_matrix(Y_test, prediction4)
        sns.heatmap(cm4, annot=True, cmap='Greens')
        plt.xlabel('Label')
        plt.ylabel('Prediction')
        plt.show()
        return print("Accuracy:", np.round(accuracy4, 2), '\n' ,classification_report(Y_test, prediction4))
    
    else:
        pass

    if LOG==True:
        logistic_model = LogisticRegression(random_state=rs)
        logistic_model.fit(X_train, Y_train)
        prediction5 = logistic_model.predict(X_test)
        accuracy5 = accuracy_score(Y_test, prediction5)
        cm5 = confusion_matrix(Y_test, prediction5)
        sns.heatmap(cm5, annot=True, cmap='PuRd')
        plt.xlabel('Label')
        plt.ylabel('Prediction')
        plt.show()
        return print("Accuracy:", np.round(accuracy5, 2), '\n' ,classification_report(Y_test, prediction5))
    
    else:
        pass

```

```python
classifier(DT=True)
```

```python
classifier(RF=True)
```

```python
classifier(SVM=True) # 가장 좋은 성능을 보임
```

```python
classifier(SGD=True)
```

```python
classifier(LOG=True)
```
